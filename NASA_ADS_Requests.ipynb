```python
def get_total(year,token):
    r = requests.get("https://api.adsabs.harvard.edu/v1/search/query?q=year%3A"+ str(year) +"&start=&rows=1",\
                headers={'Authorization': 'Bearer ' + token})
    return r.json()['response']['numFound']
    
def nasa_ads_scraper(start,year,token):
    URL = "https://api.adsabs.harvard.edu/v1/search/query?q=year%3A"+ str(year) +"&fl=abstract+%2C+ack+%2C+aff+%2C+aff_id+%2C+alternate_bibcode+%2C+alternate_title+%2C+arxiv_class+%2C+author+%2C+author_count+%2C+author_norm+%2C+bibcode+%2C+bibgroup+%2C+bibstem+%2C+citation+%2C+citation_count+%2C+cite_read_boost+%2C+classic_factor+%2C+comment+%2C+copyright+%2C+data+%2C+database+%2C+date+%2C+doctype+%2C+doi+%2C+eid+%2C+entdate+%2C+entry_date+%2C+resources+%2C+facility+%2C+first_author+%2C+first_author_norm+%2C+grant+%2C+grant_agencies+%2C+grant_id+%2C+id+%2C+identifier+%2C+indexstamp+%2C+inst+%2C+isbn+%2C+issn+%2C+issue+%2C+keyword+%2C+keyword_norm+%2C+keyword_schema+%2C+lang+%2C+links_data+%2C+nedid+%2C+nedtype+%2C+orcid_pub+%2C+orcid_other+%2C+orcid_user+%2C+page+%2C+page_count+%2C+page_range+%2C+property+%2C+pub+%2C+pub_raw+%2C+pubdate+%2C+pubnote+%2C+read_count+%2C+reference+%2C+simbid+%2C+title+%2C+vizier+%2C+volume+%2C+year&start="+ str(start) +"&rows=2000&sort=id+asc&"
    r = requests.get(URL,\
            headers={'Authorization': 'Bearer ' + token})
    return r#.json()
```
